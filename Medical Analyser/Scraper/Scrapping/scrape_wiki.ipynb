```python
from bs4.element import Comment 
def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True

texts = soup2.findAll(text=True)
visible_texts = filter(tag_visible, texts)  
print(u" ".join(t.strip() for t in visible_texts))
```

```python
  
def traverse(element):
    s=""
    if type(element) is type(list(soup2.children)[3]):
        flag = 1
        for i in list(element.children):
            if i is type(list(soup2.children)[3]):
                flag = 0   
                break
        if flag==1:
            return "<"+element.name+">"+element.text+"</"+element.name+">" 
        s=s+"<"+element.name+">"
        for i in list(element.children):
            s=s+traverse(i)
        s=s+"</"+element.name+">" 
        return s
    else:
        return s

html=traverse(list(soup2.children)[3])  
```

```python
from time import sleep
import urllib.parse as urlparse
url1='https://google.com/search?q='
l=[]
for i in range(149):
    dis=df['Source'].loc[i]
    wordlist=dis.split(' ')
    url=url1
    for w in wordlist:
        if w!=' ':
            url=url+w+'+'
    url=url+'wikipedia'
    pages = requests.get(url)
    soup = BeautifulSoup(pages.content,'html.parser')
    url2=''
    for r in soup.find_all(class_='r'):
        newurl = list(r.children)[0]['href']
        if re.search("en.wikipedia.org",newurl):
            data = urlparse.parse_qs(urlparse.urlparse(newurl).query)
            url2 = data['q'][0]
            #print(url2)
            #url2 = newurl[re.search("https:",newurl).start():re.search("&sa",newurl).start()]
            break
    if url2 == '':
        continue
    pages2 = requests.get(url2)
    soup2 = BeautifulSoup(pages2.content,'html.parser')
    sentence = ''
    parent = soup2.find('div',class_='mw-parser-output')
    flag = 0
    count = 0
    for ch in list(parent.children):
        if type(parent) is type(ch):
            if ch.name == 'h2' and (re.search('prevention', ch.span.text, re.IGNORECASE) or re.search('treatment', ch.span.text, re.IGNORECASE) or re.search('research', ch.span.text, re.IGNORECASE) or re.search('management', ch.span.text, re.IGNORECASE)):
                flag = 1
                count = count +1
                #print(ch.text)
            elif ch.name == 'h2':
                flag = 0
        if flag == 1:
            if type(parent) is type(ch):
                sentence = sentence+"<"+ch.name+">"+ch.text+"</"+ch.name+">"
    innerl=[]
    innerl.append(df.iloc[i][1])
    innerl.append(sentence)
    l.append(innerl)
    print(i)
    if count == 0:
        notfound.append(i)
    sleep(2)
```
